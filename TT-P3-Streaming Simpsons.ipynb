# Limpieza de datos en el df

```python
# Mostrar los nombres de las columnas
df.columns
```

```python
# Bucle que itera sobre los encabezados y los pone todos en minúsculas
column_names_lowered=[]
for header in df.columns:
    lower_name=header.lower()
    column_names_lowered.append(lower_name)
df.columns=column_names_lowered
df.columns
```

```python
# Bucle que itera sobre los encabezados y elimina los espacios
column_no_space=[]
for header in df.columns:
    no_space_name=header.strip()
    column_no_space.append(no_space_name)
df.columns=column_no_space
df.columns
```

```python
# Cambia el nombre de la columna "userid" para seguir la regla de snake_case
df.rename(columns={"userid": "user_id"}, inplace=True)

#Comprobar los resultados
df.columns

```

```python
#Calcular el número de valores ausentes
df.isna().sum()
```

```python
#Bucle en los encabezados para reemplazar los valores ausentes con 'unknown'
columns_to_replace=['track', 'artist', 'genre']
for column in columns_to_replace:
    df[column] = df[column].fillna('unknown')

#Contar los valores ausentes
df.isna().sum()
```

```python
#Contar los duplicados explícitos
df.duplicated().sum()
```

```python
#Eliminar duplicados explícitos
df = df.drop_duplicates()
```

```python
#Comprobar de nuevo si hay duplicados en el df
df.duplicated().sum()
```

```python
#Inspeccionar los nombres de géneros únicos
df_genres=df["genre"].unique()
print(sorted(df_genres))

```

```python
#Definir función para reemplazar los duplicados implícitos
def replace_wrong_values(df,column,wrong_values,correct_value):
    for col in df:
        df[col].replace(wrong_values, correct_value)

```

```python
#Usar la función paea eliminar los duplicados implícitos
replace_wrong_values(df,["genre"],["hip","hop","hip-hop"],"hiphop")
```

```python
#Comprobar de nuevo los duplicados implícitos
df_genres=df["genre"].unique()
print(sorted(df_genres))
```

```python
#Calcular las canciones reproducidas lunes y viernes
df_by_days=df.groupby('day')['track'].count()
print(df_by_days.loc[["Monday","Friday"]])

```

El viernes se reprodujeron más canciones que el lunes.

```python
#Definir la función number_tracks() con dos parámetros: day= y city=
def number_tracks(day,city):

    #Almacenar las filas del DataFrame donde el valor en la columna 'day' es igual al parámetro day=
    filter_by_day=df[df['day'] == day]
    #Filtrar las filas donde el valor en la columna 'city' es igual al parámetro city=
    filter_by_city=filter_by_day[filter_by_day['city'] == city]
    #Extraer la columna 'user_id' de la tabla filtrada y aplica el método count()
    result=filter_by_city["user_id"].count()
    #Devuelver el número de valores de la columna 'user_id'
    return result
```

```python
#Llamar a la función para determinar el número de canciones reproducidas en Springfield el lunes
print(number_tracks("Monday","Springfield"))
```

```python
#Determinar el número de canciones reproducidas en Shelbyville el lunes
print(number_tracks("Monday","Shelbyville"))

```

```python
#El número de canciones reproducidas en Springfield el viernes
print(number_tracks("Friday","Springfield"))

```

```python
# El número de canciones reproducidas en Shelbyville el viernes
print(number_tracks("Friday","Shelbyville"))
```
